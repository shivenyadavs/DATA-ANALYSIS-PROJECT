```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df=pd.read_csv("dap_final.csv")
df
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Calculate average profit per payment method
avg_profit = df.groupby('Payment')['Profit'].mean().reset_index()

# Plotting the grouped bar chart
plt.figure(figsize=(10, 6))
plt.bar(avg_profit['Payment'], avg_profit['Profit'], color='skyblue',linewidth=5,width=0.3)
plt.xlabel('Payment Method')
plt.ylabel('Average Profit')
plt.title('Average Profit by Payment Method')
plt.show()
```

```python
Here we are comparing the average profit associated with different payment methods to identify if certain payment methods are associated with higher or lower profits.
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Calculate average profit per offer type
avg_profit = df.groupby('Offers')['Profit'].mean().reset_index()

# Plotting the line chart
plt.figure(figsize=(10, 6))
plt.plot(avg_profit['Offers'], avg_profit['Profit'], marker='o', color='blue', linestyle='-', linewidth=2)
plt.xlabel('Offers')
plt.ylabel('Average Profit')
plt.title('Average Profit vs Offers')
plt.grid(True)
plt.show()
```

```python
# Create a pivot table for heatmap
pivot_data = df.pivot_table(index='Product line', columns='City', values='Profit', aggfunc='sum')

# Plotting the heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(pivot_data, cmap='YlGnBu', annot=True, fmt=".0f")
plt.xlabel('City')
plt.ylabel('Product Line')
plt.title('Profit Comparison Across Product Lines and City')
plt.show()
```
