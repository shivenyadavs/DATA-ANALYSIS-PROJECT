```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("Mobile_Dataset.csv")
df
```

what is the best representation for model name and specs score?

```python
plt.figure(figsize=(10,11))
plt.bar(df['Name'],df['Specs_Score'])
plt.xlabel('name')
plt.ylabel('specs_score')
plt.xticks (rotation='vertical')
plt.legend()
plt.show()
```

Reason You have a categorical variable on one axis and want to visualize the corresponding numerical values.

what is the best representation for model name and display?

```python
plt.figure(figsize=(10,11))
plt.plot(df['Name'],df['display'])
plt.ylabel('display')
plt.xlabel('Name')
plt.xticks (rotation='vertical')
plt.legend()
plt.show()
```

Reason line plot represents the trend or changes in display size for each model name

how to measure the distribution or spread of data points using price and main camera?

```python
plt.figure(figsize=(10,9))
plt.scatter(df['Price(Rs.)'],df['Main_Camera'],marker='o',color='b')
plt.ylabel('Main_Camera')
plt.xlabel('Price(Rs.)')
plt.xticks (rotation='vertical')
plt.grid(True)
plt.legend()
plt.show()
```

Reason scatter plot spots data points that deviates from general data points help in finding the errors or unusal changes 

what is the best measure for visualizing the change in distribution between the RAM and Battery?

```python
plt.figure(figsize=(10,9))
plt.scatter(df['RAM'],df['Battery'])
plt.ylabel('Battery')
plt.xlabel('RAM')
plt.xticks (rotation='vertical')
plt.grid(True)
plt.legend()
plt.show()
```

Reason scatter plot is used here to find the data point distribution from general data points which indicates outliers and help in pattern recognition

what is the best measure to find the pictorail relationship between categorical data and numerical data?

```python
plt.figure(figsize=(10,11))
plt.bar(df['Name'],df['Price(Rs.)'])
plt.xlabel('name')
plt.ylabel('Price(Rs.)')
plt.xticks (rotation='vertical')
plt.legend()
plt.show()
```

Reason bar graph is the best measure to be used here because it allows comparison between numerical value associated with the other categorical data

what is the best measure of finding outliers or distributions between categorical datasets?

```python
plt.boxplot(df['display'])
plt.xlabel('column 7')
plt.ylabel('Display')
plt.show()
```

Reason boxplot gives clear representation of central tendency(median),spread,outliers. it is easy to find the larger display or smaller display with the model name.

what is the best measure to compare batteries and across different brand name?

```python
plt.boxplot(df['Battery'])
plt.xlabel('column 8 ')
plt.ylabel('Battery')
plt.show()
```

Reason it is easy to identify brands with larger,median and smaller battery capacities as well as brands with more or less variabilities in battery capacities it als gives outliers with high or low batteries capcities
